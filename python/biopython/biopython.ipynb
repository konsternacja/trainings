#### Reading fasta file

```python
import cProfile
import timeit

from Bio import SeqIO
```

```python
def process_sequences1():
    seq_record = SeqIO.parse('example.fasta', 'fasta')

    for sequence in seq_record:
        print(sequence.id, sequence.seq)
        
cProfile.run('process_sequences1()')
```

```python
def process_sequences2():
    for sequence in SeqIO.parse('example.fasta', 'fasta'):
        return sequence.id, sequence.seq

cProfile.run('process_sequences2()')
```

```python
time1 = timeit.timeit(process_sequences1, number=1000)
time2 = timeit.timeit(process_sequences2, number=1000)

print(f"Time taken by function1: {time1} seconds")
print(f"Time taken by function2: {time2} seconds")
```

There was not much of a difference but it was better not to create additional variable

#### Central Dogma

```python
from Bio.Seq import Seq
```

```python
# Define a DNA sequence
seq = Seq("ATATGCGATGCTGCAAACTG")

# Transcribe, complement, and translate the sequence
transcribed_seq = seq.transcribe()
complement_seq = seq.complement()
translated_seq = seq.translate()

# Print the results
print(f"Transcribed seq: {transcribed_seq}, Complement seq: {complement_seq}, Translated seq: {translated_seq}.")
```
