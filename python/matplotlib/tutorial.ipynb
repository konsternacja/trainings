## Basic graphs

### Line plot

```python
x = [1,2,3]
y=[2,4,6]

plt.plot(x,y)
plt.show()
```

```python
x = [0,1,2,3,4]
y=[0,2,4,6,8]

# resize (dpi - pixels per inch)
plt.figure(figsize=(7,5), dpi=100)

# plt.plot(x,y, label='2x', color = 'red', marker = 'x', markersize = 10, markeredgecolor = 'blue', linewidth = 1.2, linestyle='--')
fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--',label='2x', )

# adding second line
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')
plt.plot(x2[5:], x2[5:]**2, 'r--')

# adding title and labels
plt.title('Basic Graph', fontdict={'fontname': 'serif', 'fontsize': 20})
plt.xlabel('X Axis', fontdict={'fontname': 'serif', 'fontsize': 15})
plt.ylabel('Y Axis', fontdict = {'fontname': 'serif', 'fontsize': 15})

# resizing
plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])

# adding legend
plt.legend()

# saving a plot
plt.savefig('graph1.png', dpi=300)

plt.show()
```

### Bar chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

# first labels the values!
bars = plt.bar(labels, values, color = 'green')

# hatch (really cool)
# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

pattern = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(pattern.pop(0))

plt.figure(figsize=(7,5), dpi=100)

plt.show()
```

### Pie charts

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['pink', 'purple']

plt.pie([left, right], labels=labels, colors=colors, autopct='%.2f')
plt.title('Foot Preference of FIFA Players', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.show()
```

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```

In piecharts you firstly add weights, then labels (for example ['left', 'right']), then the rest

### Box and Whiskers 

```python
fifa.columns
```

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall'] # we get column overall for fc barcelona rows
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']: # tak odwoujesz si do poszczeglnych bars w wykresie wykres['boxes]
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    box.set(hatch = '/')
    
plt.show()
```
