```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

```python
# Distribution plot

# kde=False we don't have line
sns.displot(crash_df['not_distracted'], kde=False, bins=25)
```

```python
# joint plot

# we compare 2 distributions
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
```

```python
# kde plot

sns.kdeplot(crash_df['alcohol'])
```

```python
# pair plot

# relationships across the entire data frame
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
sns.pairplot(tips_df, hue='sex', palette='coolwarm')
```

```python
# rug plot

sns.rugplot(tips_df['tip'])
```

```python
# styling for axis and grids

sns.set_style('white')
# whitegrid, dark, ticks etc.

sns.set_context('paper', font_scale=1.4)

plt.figure(figsize=(8,5))
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')

# no lines for borders
sns.despine(left=True, bottom=True, top=True, right=True)
```

```python
# bar plot

sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.cov)
```

```python
# count plot

sns.countplot(x='sex', data=tips_df, palette='coolwarm')
```

```python
# box plot

sns.set_style('whitegrid')
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')

plt.legend(loc=1)
```

```python
# violin plot

sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='coolwarm', split=True)

plt.legend(loc=0)
plt.ylabel('Total Bill')

plt.show()
```

```python
# strip plots

plt.figure(figsize=(16,10))

# dodge - separate the hue
sns.stripplot(x='day', y='total_bill', data=tips_df, 
              jitter=True, hue='sex', palette='Set1', dodge=True)

plt.legend(loc=0)
plt.show()
```

```python
# swarm plot

sns.violinplot(x='day', y='total_bill', data=tips_df)
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='black')
```

```python
# palettes

plt.figure(figsize=(16,10))

sns.set_style('whitegrid')
sns.set_context('talk')
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='Accent')

# magma, inferno, viridis, coolwarm, Set1, Set2, Set3, Pastel1, Pastel2, Dark2, Accent, Paired, Deep, muted, bright, colorblind
```

```python
# heatmaps

plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)

crash_mx = crash_df.drop('abbrev', axis=1).corr()
crash_mx

sns.heatmap(crash_mx, annot=True, cmap='Blues')
```

```python
iris = sns.load_dataset('iris')
```

```python
# pair grids

iris_g = sns.PairGrid(iris, hue="species")

iris_g.map(plt.scatter)

iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter)

iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
```

```python
tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.3,
                      col_order=['Dinner', 'Lunch'], palette='Set1')
tips_fg.map(plt.scatter, "total_bill", "tip", edgecolor='w')
```

```python
kws = dict(s=50, linewidth=.5, edgecolor='w')
tips_fg = sns.FacetGrid(tips_df, col='sex', hue='smoker', height = 4, aspect=1.3, 
                        hue_order=['Yes','No'], hue_kws=dict(marker=['^','v']))
tips_fg.map(plt.scatter, 'total_bill', 'tip', **kws)
```
